```python
from datetime import datetime
def greet_with_time(name):
    current_time = datetime.now().strftime("%H:%M:%S")
    print(f"Hello {name}, current time is {current_time}")
greet_with_time("Saad")
```

```python
def calculate_average(num1, num2, num3):
    return (num1 + num2 + num3) / 3
average = calculate_average(10, 20, 30)
print("Average:", average)
```

```python
def calculate_discount(price, discount_percentage=10):
    discount = price * discount_percentage / 100
    final_price = price - discount
    return final_price
print("Final Price:", calculate_discount(1000))
print("Final Price with 20% discount:", calculate_discount(1000, 20))
```

```python
def print_info(name, age, city):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)
print_info(name="Saad", age=21, city="Lahore")
```

```python
def find_max(*numbers):
    return max(numbers)
print("Maximum Value:", find_max(5, 10, 15, 2, 30))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
print("Factorial:", factorial(5))
```

```python
def apply_function(func, numbers):
    result = []
    for num in numbers:
        result.append(func(num))
    return result
def square(x):
    return x * x
numbers = [1, 2, 3, 4, 5]
print("Squared List:", apply_function(square, numbers))
```
