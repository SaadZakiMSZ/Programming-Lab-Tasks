```python
orders = [
    {"product": "shirt", "category": "clothing"},
    {"product": "jeans", "category": "clothing"},
    {"product": "laptop", "category": "electronics"},
    {"product": "phone", "category": "electronics"},
    {"product": "t-shirt", "category": "clothing"}
]

category_count = {}

for order in orders:
    cat = order["category"]

    if cat in category_count:
        category_count[cat] += 1
    else:
        category_count[cat] = 1

popular_category = max(category_count, key=category_count.get)

print(category_count)
print(popular_category)
```

```python
inventory = {}

def add_product(name, quantity, price):
    inventory[name] = {
        "quantity": quantity,
        "price": price
    }
def update_quantity(name, quantity):
    if name in inventory:
        inventory[name]["quantity"] = quantity
def total_value():
    total = 0
    for item in inventory:
        total += inventory[item]["quantity"] * inventory[item]["price"]
    return total
def low_stock(threshold):
    low_items = []
    for item in inventory:
        if inventory[item]["quantity"] < threshold:
            low_items.append(item)
    return low_items
add_product("apple", 50, 2)
add_product("banana", 20, 1)
add_product("mango", 5, 5)
update_quantity("banana", 10)
print(inventory)
print(total_value())
print(low_stock(15))
```

```python
orders = [
    {"product": "shirt", "category": "clothing"},
    {"product": "jeans", "category": "clothing"},
    {"product": "laptop", "category": "electronics"},
    {"product": "phone", "category": "electronics"},
    {"product": "t-shirt", "category": "clothing"}
]
category_count = {}
for order in orders:
    cat = order["category"]
    if cat in category_count:
        category_count[cat] += 1
    else:
        category_count[cat] = 1
popular_category = max(category_count, key=category_count.get)
print(category_count)
print(popular_category)
```

```python
store = {
    "notebook": {"price": 150, "quantity": 40},
    "marker": {"price": 60, "quantity": 80},
    "calculator": {"price": 1200, "quantity": 15}
}
cart = {}
def add_item(item, qty):
    if item in store and store[item]["quantity"] >= qty:
        cart[item] = cart.get(item, 0) + qty
        store[item]["quantity"] -= qty
def remove_item(item, qty):
    if item in cart:
        cart[item] -= qty
        if cart[item] <= 0:
            del cart[item]
def calculate_total():
    total = 0
    for item, qty in cart.items():
        total += store[item]["price"] * qty
    return total
def final_amount(total, discount_rate, tax_rate):
    discounted = total - (total * discount_rate)
    taxed = discounted + (discounted * tax_rate)
    return taxed
add_item("notebook", 3)
add_item("calculator", 1)
cost = calculate_total()
final_cost = final_amount(cost, 0.15, 0.08)
print(cart)
print(cost)
print(final_cost)
```
